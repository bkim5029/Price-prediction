### P value
P value is the probability value that the correlation between these two variables is significant.
- p value of 0.05 means we are 95% confident that the correlation is significant
- generally anything below 0.05 is what we are looking for

```python
from scipy import stats
```

# ANOVA (Analysis of Variance)
Anova test does not tell you the type of relationship that exists but only that relationship exists.
- ANOVA is used to test whether there are significant differences between the means of two or more groups.
- F value is calculated (Variation between group of data / Variation within its own data)

<p><b>data1</b> = [13,43,24,48,20]

<p><b>data2</b> = [12,42,21,45,21]

- data1 has `High` variation within its group but `Low` variation between data2.
    
<p><b>data3</b> = [1,2,3,5,6]
    
<p><b>data4</b> = [10,11,13,14,15]
   
- data3 has `Low` variation within its group but `High` variation between data4
    
The main idea of ANOVA is to figure out how much total variance comes from
- if most of the variation is `between groups(High F-value)`, the data does have a significant effect
- if most of the variation is `within groups(Low F-value)`, the data does not have a significant effect

<p><b>F-test score</b>: calculates how much the actual means deviate from the assumption, and reports it as the   F-test score. A larger score means there is a larger difference between the means which means we reject the null hypothesis.</p>

`F = (Variance BETWEEN groups / Variance WITHIN groups)`

<p><b>P-value</b>:  P-value tells how statistically significant our calculated score value is.</p>

### Large F value means there is a high probability that the groups do have different means meaning the data does make a difference (Rejecting Null Hypothesis)
